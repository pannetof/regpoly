# Tausworthe Generator Equidistribution

Tausworthe generators defined by a trinomial $z^k + z^q + 1$ over GF(2)
with decimation step $s$. We test several generators with different $k$ values.

```python
from regpoly.generateur import Generateur
from regpoly.combinaison import Combinaison
from regpoly.analyses.equidistribution_test import (
    EquidistributionTest, METHOD_MATRICIAL
)

INT_MAX = 2**31 - 1
```

## 1. Create Tausworthe Generators

Parameters: `poly` (list of exponents of the primitive polynomial, e.g. `[31, 3]` for $z^{31} + z^3 + 1$),
`s` (decimation step).

```python
# Tausworthe generators with primitive trinomials z^k + z^q + 1
# poly = [k, q] defines the polynomial; s is the decimation step
taus_configs = [
    {"poly": [31, 3],  "s": 12, "label": "z^31 + z^3 + 1"},
    {"poly": [61, 5],  "s": 24, "label": "z^61 + z^5 + 1"},
    {"poly": [89, 38], "s": 6,  "label": "z^89 + z^38 + 1"},
]

generators = []
for cfg in taus_configs:
    k = max(cfg["poly"])
    L = min(32, k)
    g = Generateur.create("Tausworthe", L, poly=cfg["poly"], s=cfg["s"])
    generators.append((g, cfg))
    print(f"Tausworthe {cfg['label']}, s={cfg['s']} -> k={g.k}, L={g.L}")
```

## 2. Equidistribution for Each Generator

```python
for g, cfg in generators:
    comb = Combinaison(J=1, Lmax=g.L)
    comb.components[0].add_gen(g)
    comb.reset()

    test = EquidistributionTest(L=g.L, delta=[INT_MAX]*(g.L+1),
                                mse=INT_MAX, method=METHOD_MATRICIAL)
    result = test.run(comb)

    print(f"\n=== Tausworthe {cfg['label']}, s={cfg['s']} ===")
    print(f"Dimension gaps sum = {result.se}")
    print(result.display_table(comb, 'l')[0])
```

## 3. Combined Tausworthe Generators

Combine k=31 and k=61 generators (k_total = 92).

```python
g1, _ = generators[0]
g2, _ = generators[1]
L_comb = min(g1.L, g2.L)

comb = Combinaison(J=2, Lmax=L_comb)
comb.components[0].add_gen(g1)
comb.components[1].add_gen(g2)
comb.reset()

print(f"Combined: k_g={comb.k_g}, L={comb.L}")

test = EquidistributionTest(L=comb.L, delta=[INT_MAX]*(comb.L+1),
                            mse=INT_MAX, method=METHOD_MATRICIAL)
result = test.run(comb)

print(f"Dimension gaps sum = {result.se}")
print(result.display_table(comb, 'l')[0])
```
