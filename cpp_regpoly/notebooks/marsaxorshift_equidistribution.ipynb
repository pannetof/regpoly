# Marsaglia Xor-shift Equidistribution

Marsaglia's Xor-shift generators of types 1, 3, and 4.
Each type uses a different combination of shift-and-XOR operations.

```python
from regpoly.generateur import Generateur
from regpoly.combinaison import Combinaison
from regpoly.analyses.equidistribution_test import (
    EquidistributionTest, METHOD_MATRICIAL
)

INT_MAX = 2**31 - 1
```

## 1. Type 1 --- Single-word triple Xor-shift

$x \leftarrow x \oplus (x \ll a)$, then $\oplus (x \gg b)$, then $\oplus (x \ll c)$.
State size: $k = w = 32$.

```python
gen_t1 = Generateur.create("MarsaXorshiftGen", 32, type=1, w=32, shifts=[13, -17, 5])

print(f"{gen_t1.name()}: k={gen_t1.k}, L={gen_t1.L}")
gen_t1.display()

comb = Combinaison(J=1, Lmax=32)
comb.components[0].add_gen(gen_t1)
comb.reset()

test = EquidistributionTest(L=32, delta=[INT_MAX]*33,
                            mse=INT_MAX, method=METHOD_MATRICIAL)
result = test.run(comb)
print(f"\nDimension gaps sum = {result.se}")
result.display_table(comb, 'l')
```

## 2. Type 3 --- Multi-tap Xor-shift

Multiple tap positions in a circular buffer, each with its own shift.
State size: $k = w \times r$.

```python
# Type 3, r=2, 7 taps
gen_t3 = Generateur.create("MarsaXorshiftGen", 32,
    type=3, w=32, r=2,
    tap_positions=[1, 1, 2, 1, 2, 2, 1],
    tap_shifts=[7, -18, -15, 9, 18, -8, -20])

print(f"{gen_t3.name()}: k={gen_t3.k}, L={gen_t3.L}")
gen_t3.display()

comb = Combinaison(J=1, Lmax=32)
comb.components[0].add_gen(gen_t3)
comb.reset()

result = test.run(comb)
print(f"\nDimension gaps sum = {result.se}")
result.display_table(comb, 'l')
```

## 3. Type 4 --- Two-component, 2 shifts each

$x \leftarrow x \oplus S^{p_0}(x) \oplus S^{p_1}(x)$,
$y \leftarrow y \oplus S^{q_0}(y) \oplus S^{q_1}(y)$,
output = $x \oplus y$. State size: $k = w \times r$.

```python
# Type 4, r=8, m=1
gen_t4 = Generateur.create("MarsaXorshiftGen", 32,
    type=4, w=32, r=8, m=1,
    p=[-31, -25], q=[-7, 3])

print(f"{gen_t4.name()}: k={gen_t4.k}, L={gen_t4.L}")
gen_t4.display()

comb = Combinaison(J=1, Lmax=32)
comb.components[0].add_gen(gen_t4)
comb.reset()

result = test.run(comb)
print(f"\nDimension gaps sum = {result.se}")
result.display_table(comb, 'l')
```
