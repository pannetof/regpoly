# GF(2^w) Generator Equidistribution

Generators operating in the extension field GF($2^w$).
Two variants: Polynomial LCG and LFSR over GF($2^w$).

```python
import regpoly._regpoly_cpp as _cpp
from regpoly.generateur import Generateur
from regpoly.combinaison import Combinaison
from regpoly.analyses.equidistribution_test import (
    EquidistributionTest, METHOD_MATRICIAL
)

INT_MAX = 2**31 - 1
```

## 1. GF(2^w) Polynomial LCG

Parameters: `w` (extension field degree), `r` (number of registers),
coefficients defining the recurrence over GF($2^w$).
State size: $k = w \times r$.

```python
# GF(2^7) PolyLCG with r=3, k=21
gen_poly = Generateur(_cpp.create_generator("GenF2wPolyLCG", {
    "w": 7, "r": 3, "step": 49,
    "coeffs": [
        {"value": 0x3c, "position": 2},
        {"value": 0x1c, "position": 1},
    ],
    "modM": 0,
    "normal_basis": False
}, 7))

print(f"{gen_poly.name()}: k={gen_poly.k}, L={gen_poly.L}")
gen_poly.display()

comb = Combinaison(J=1, Lmax=gen_poly.L)
comb.components[0].add_gen(gen_poly)
comb.reset()

test = EquidistributionTest(L=gen_poly.L, delta=[INT_MAX]*(gen_poly.L+1),
                            mse=INT_MAX, method=METHOD_MATRICIAL)
result = test.run(comb)
print(f"\nDimension gaps sum = {result.se}")
result.display_table(comb, 'l')
```

## 2. GF(2^w) LFSR

```python
# GF(2^7) LFSR with r=3, k=21
gen_lfsr = Generateur(_cpp.create_generator("GenF2wLFSR", {
    "w": 7, "r": 3, "step": 49,
    "coeffs": [
        {"value": 0x3c, "position": 2},
        {"value": 0x1c, "position": 1},
    ],
    "modM": 0,
    "normal_basis": False
}, 7))

print(f"{gen_lfsr.name()}: k={gen_lfsr.k}, L={gen_lfsr.L}")
gen_lfsr.display()

comb = Combinaison(J=1, Lmax=gen_lfsr.L)
comb.components[0].add_gen(gen_lfsr)
comb.reset()

result = test.run(comb)
print(f"\nDimension gaps sum = {result.se}")
result.display_table(comb, 'l')
```
