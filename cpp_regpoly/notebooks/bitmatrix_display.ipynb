# BitMatrix Display Formats

Showcase of the six display formats for the transition matrix of a
PolyLCG generator ($k=32$, polynomial $z^{32}+z^{30}+z^{25}+z^{24}+z^{22}+z^{15}+z^6+z^2+1$).

```python
from regpoly.generateur import Generateur
from regpoly.matrix import BitMatrix
```

## Create the generator and its transition matrix

```python
# First polynomial from 32poly.dat: z^32 + z^30 + z^25 + z^24 + z^22 + z^15 + z^6 + z^2 + 1
# Exponents (excluding z^32 leading term): [30, 25, 24, 22, 15, 6, 2, 0]
gen = Generateur.create("PolyLCG", 32, k=32, poly=[30, 25, 24, 22, 15, 6, 2, 0])

print(f"Generator: {gen.name()}, k={gen.k}")
print(gen.display())

A = gen.transition_matrix()
print(f"\nTransition matrix: {A.nblignes} x {A.l}")
```

## 1. Raw binary (0/1)

```python
print(A.display("raw"))
```

## 2. Hexadecimal (4-bit groups)

```python
print(A.display("hex4"))
```

## 3. Hexadecimal (8-bit groups)

```python
print(A.display("hex8"))
```

## 4. Braille characters

Each Unicode braille character encodes a 2-column × 4-row block,
giving a very compact representation.

```python
print(A.display("braille"))
```

## 5. Block characters

Each row of output represents two matrix rows using
half-block characters (▀ ▄ █).

```python
print(A.display("block"))
```

## 6. PNG bitmap

Rendered as a PIL image: black = 1, white = 0.

```python
A.to_image(scale=12)
```

## Bonus: larger matrix (TGFSR k=96)

```python
gen96 = Generateur.create("TGFSRGen", 32, w=32, r=3, m=1, a=0x9965523b)

B = gen96.transition_matrix()
print(f"TGFSR transition matrix: {B.nblignes} x {B.l}")
print()
print(B.display("braille"))
```

```python
B.to_image(scale=6)
```
