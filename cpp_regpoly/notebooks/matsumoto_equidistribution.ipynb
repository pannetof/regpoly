# Matsumoto Generator Equidistribution

Matsumoto-type generators with two variants:
- Type 1: basic recurrence with shift parameters
- Type 2: recurrence with an additional twist matrix coefficient $a$

```python
from regpoly.generateur import Generateur
from regpoly.combinaison import Combinaison
from regpoly.analyses.equidistribution_test import (
    EquidistributionTest, METHOD_MATRICIAL
)

INT_MAX = 2**31 - 1
```

## 1. Matsumoto Type 1

Parameters: `n` (number of words), `m` (middle distance),
and integer shift parameters. State size: $k = 32 \times n$.

```python
gen_t1 = Generateur.create("matsumoto", {
    "type": 1, "n": 3, "m": 1,
    "paramsint": [4, 5, -7, 11, -15],
    "paramsunsigned": [0]
}, 32))

print(f"{gen_t1.name()}: k={gen_t1.k}, L={gen_t1.L}")
gen_t1.display()

comb = Combinaison(J=1, Lmax=32)
comb.components[0].add_gen(gen_t1)
comb.reset()

test = EquidistributionTest(L=32, delta=[INT_MAX]*33,
                            mse=INT_MAX, method=METHOD_MATRICIAL)
result = test.run(comb)
print(f"\nDimension gaps sum = {result.se}")
result.display_table(comb, 'l')
```

## 2. Matsumoto Type 2

Same recurrence structure but with a twist matrix coefficient $a$.

```python
gen_t2 = Generateur.create("matsumoto", {
    "type": 2, "n": 3, "m": 1,
    "paramsint": [4, 5, -7, 11, -15],
    "paramsunsigned": [1, 0x9908b0df]
}, 32))

print(f"{gen_t2.name()}: k={gen_t2.k}, L={gen_t2.L}")
gen_t2.display()

comb = Combinaison(J=1, Lmax=32)
comb.components[0].add_gen(gen_t2)
comb.reset()

result = test.run(comb)
print(f"\nDimension gaps sum = {result.se}")
result.display_table(comb, 'l')
```
